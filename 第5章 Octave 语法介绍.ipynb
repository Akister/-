## 数据操作
A(3,2) $\quad$  %取出矩阵A中第3行第2列的元素

A(2,:)  $\quad$ %取出矩阵A中第2行中所有元素

A\*B  $\quad$    %矩阵乘法

A.\*C  $\quad$   %两个矩阵对应位置元素进行相乘（两个矩阵维度相同）

A.^2  $\quad$   %矩阵A的每个元素进行平方

1./A   $\quad$  %矩阵A的每个元素的倒数

log(A) $\quad$  %对矩阵A的每个元素求自然对数（以e为底）

exp(A) $\quad$  %求自然对数e以矩阵A的元素次方

abs(V) $\quad$  %求矩阵V的元素的绝对值

-V   $\quad$    %求矩阵V的元素的相反数（等效于-1×V）

A'   $\quad$    %矩阵转置

max(a) $\quad$  %求向量a中的最大值，同min(a)

max(A,[],1) $\quad$  %求矩阵A中每一列的最大值，返回1行×A的列数矩阵，效果等同于max(A),同min

max(A,[],2) $\quad$ %求矩阵A中每一行的最大值，返回A的行数×1矩阵，同min

A\<4   $\quad$   %将矩阵或向量A中的值与4比较，相应元素位置返回相应元素比较的逻辑结果

find(A\<3) %返回矩阵A中元素值\<3的元素的下标，find(A)则返回矩阵A中的元素值非0的元素的下标

magic(3) $\quad$ %快速产生一个3维的魔法矩阵（每行每列美对角线每分块的数字相加之和相等），最低维度3维

sum(A)  $\quad$ %若A为一维向量则返回A中所有元素的和（一个数字），若A为矩阵则返回一个A的列数×1的矩阵，这个矩阵的每个值为A中该列所有元素的和,可sum(A,1)/sum(A,2)

prod(A) $\quad$ %若A为一维向量则返回A中所有元素的乘积（一个数字），若A为矩阵则返回一个A的列数×1的矩阵，这个矩阵的每个值为A中该列所有元素的乘积,可prod(A,1)/prod(A,2)

floor(A) $\quad$ %将A的所有元素进行向下取整

ceil(A) $\quad$ %将A的所有元素进行向上取整

flipud(A) $\quad$ %将矩阵A上下翻转

pinv(A) $\quad$  %对矩阵A求逆，若A不存在逆矩阵，则求得A的逆矩阵

## 画图与数据可视化
plot(t,y) $\quad$ %画出关于函数y关于可变序列t的图形

eg:

t=[0:0.001:0.98];

x=sin(2\*pi\*t);

y=cos(2\*pi\*t);

plot(t,x);

hold on;  $\quad$ %将多幅图形画在一张图上

plot(t,y,'r'); $\quad$ %以红色线条画出y的图形

xlabel('sin'); $\quad$ %给X轴做注释

ylabel('cos');

legend('sin','cos'); $\quad$ %给线条做出图例

title('plot name  ;  $\quad$  %给图命名

print -dpng 'plotname.png' $\quad$  %以.png格式保存图形

#### 同时画多幅图形：
figure(1);plot(t,x);

figure(2);plot(t,y);

#### 一张图上画多幅子图（分开画）
subplot(2,3,1);  $\quad$ %将图画分成2行3列的分割，现在首先操作第1幅子图；

plot(t,x);

#### 改变坐标轴比例
axis([0.5 1 -1 1]); $\quad$  %将图形的坐标轴比例进行修改，x轴的范围为从0.5到1，y轴的范围为从-1到1;
